```python
target = 'DEATH_EVENT'
cat_names = ['anaemia', 'diabetes', 'high_blood_pressure', 'sex', 'smoking', 'RFPred']
cont_names = ['x_age', 'x_creatinine_phosphokinase', 'x_ejection_fraction', 'x_platelets', 'x_serum_creatinine', 'x_serum_sodium', 'x_time']
```

```python

from fastai.tabular import * 

procs = [FillMissing, Categorify]

test = TabularList.from_df(test, cat_names=cat_names, cont_names=cont_names, procs=procs)
```

```python
# test2 = TabularList.from_df(TEST, cat_names=cat_names, cont_names=cont_names, procs=procs)
```

```python
# Train Data Bunch
data = (TabularList.from_df(combineTrainVal, path='.', cat_names=cat_names, cont_names=cont_names, procs=procs)
                        .split_by_rand_pct()
                        .label_from_df(cols = target)
                        .databunch())
data.show_batch(rows=10)
```

```python
learn = tabular_learner(data, layers=[1000, 200, 15], metrics=accuracy, emb_drop=0.1, callback_fns=ShowGraph)

# select the appropriate learning rate
learn.lr_find()

# we typically find the point where the slope is steepest
learn.recorder.plot()

# Fit the model based on selected learning rate
learn.fit_one_cycle(15, max_lr=slice(1e-01))

# Analyse our model
learn.model
learn.recorder.plot_losses()
```
